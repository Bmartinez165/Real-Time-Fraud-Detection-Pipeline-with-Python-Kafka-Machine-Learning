```python
# First, install the required packages
!pip install fastapi pydantic

# Import the necessary modules
from fastapi import FastAPI
from pydantic import BaseModel

# Create mock functions to replace the missing modules
# Comment: Instead of importing non-existent modules, we create simple mock functions
def rule_engine(tx):
    # Simple mock implementation
    risk_score = 0
    reasons = []
    
    if tx.amount > 1000:
        risk_score += 30
        reasons.append("High transaction amount")
    
    if tx.ip_risk > 0.7:
        risk_score += 25
        reasons.append("High IP risk")
    
    return risk_score, reasons

def ml_model(tx):
    # Simple mock implementation
    # Returns a probability between 0 and 1
    risk_prob = (tx.amount / 10000) * tx.ip_risk
    return min(risk_prob, 1.0)

app = FastAPI()

class Transaction(BaseModel):
    user_id: int
    amount: float
    ip_risk: float
    device: str
    location: str

def calculate_risk(tx):
    rule_score, reasons = rule_engine(tx)
    ml_prob = ml_model(tx)

    final_score = rule_score + (ml_prob * 100)

    if final_score > 70:
        decision = "BLOCK"
    elif final_score > 40:
        decision = "REVIEW"
    else:
        decision = "ALLOW"

    return {
        "risk_score": round(final_score, 2),
        "decision": decision,
        "reasons": reasons
    }

@app.post("/score_transaction")
def score_transaction(tx: Transaction):
    return calculate_risk(tx)
```
